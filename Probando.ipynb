```python
import matplotlib.pyplot as plt

# Datos para el gráfico de barras
categorias = ['Categoría A', 'Categoría B', 'Categoría C', 'Categoría D']
valores = [10, 25, 15, 30]

# Crear el gráfico de barras
plt.figure(figsize=(8, 6))  # Tamaño de la figura (opcional)
plt.bar(categorias, valores, color='skyblue')

# Personalizar el gráfico
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Gráfico de Barras')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x (opcional)

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos para el gráfico de barras
categorias = ['Alternativa 1', 'Alternativa 2']  # Solo dos categorías
valores_set1 = [10, 25]
valores_set2 = [15, 20]
valores_set3 = [5, 15]

# Definir la posición de las barras en el eje x
x = np.arange(len(categorias))

# Ancho de las barras
ancho_barra = 0.2

# Crear el gráfico de barras agrupadas
plt.figure(figsize=(8, 6))
plt.bar(x - ancho_barra, valores_set1, ancho_barra, label='Servicio', color='skyblue')
plt.bar(x, valores_set2, ancho_barra, label='Costos', color='lightcoral')
plt.bar(x + ancho_barra, valores_set3, ancho_barra, label='Carga laboral', color='lightgreen')

# Personalizar el gráfico
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Resultados de alternativas para Pearson Vue')
plt.xticks(x, categorias)
plt.legend()

# Mostrar el gráfico
plt.show()
```
